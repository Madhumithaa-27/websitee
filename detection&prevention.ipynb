```python
import google.generativeai as genai
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# Load Abusive Content Detection Model
tokenizer = AutoTokenizer.from_pretrained("rahmas/abusive_content_identification")
model = AutoModelForSequenceClassification.from_pretrained("rahmas/abusive_content_identification")

# Set up Gemini AI
genai.configure(api_key="AIzaSyAy86LsYnQJKTQ9QDkc0YvMb9YwHYqYwCU")  # Replace with your API Key
model_gemini = genai.GenerativeModel("gemini-1.5-pro")

# Function to Detect Abuse
def detect_abuse(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)

    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits
        probabilities = torch.nn.functional.softmax(logits, dim=1)  # Convert logits to probabilities
        predicted_class = torch.argmax(probabilities, dim=1).item()
        confidence = probabilities[0][predicted_class].item()

    labels = ["Non-Abusive", "Abusive"]
    return labels[predicted_class], confidence

# Function to Generate Positive Alternative Using Gemini AI
def suggest_positive_alternative(abusive_text):
    prompt = f"""
    The following statement contains harmful language:
    "{abusive_text}"

    Your task is to rephrase it into a polite and constructive statement, while preserving the **core intent** in a non-offensive way.

    - If the statement contains **insults**, replace them with **neutral feedback**.
    - If it contains **aggressive language**, replace it with **encouraging words**.
    - If it promotes **harm**, reframe it into **concern and support**.

    Example:
    - "You are so dumb" → "I think you might need more practice with this."
    - "Kill yourself" → "I know things might be hard, but I'm here if you need support."
    - "Your work is garbage" → "I believe this could be improved with some refinements."

    Now, provide a polite and constructive alternative:
    """

    response = model_gemini.generate_content(prompt)
    return response.text

# User Input for Testing
user_input = input("Enter a comment: ")

# Detect Abuse
label, confidence = detect_abuse(user_input)

if label == "Abusive":
    print(f"\n🚨 Detected Abusive Comment (Confidence: {confidence:.2f})")
    alternative = suggest_positive_alternative(user_input)
    print(f"💡 Suggested Positive Alternative: {alternative}\n")
else:
    print("\n✅ No Abusive Content Detected!")

```
